```python
import pandas as pd
import matplotlib.pyplot as plt
from data_parser import parse_data
```

```python
data_folder = "data"
excel_path = "data/de_hc_adatbázis.xlsx"
```

```python
df_resnet, df_ecapa, df_xvector = parse_data(data_folder, excel_path)
```

```python
df_resnet
```

```python
cols = list(df_resnet.columns)
mata_cols = cols[:5]
feature_cols = cols[5:]
```

```python
# show number of samples in each type
df_resnet.groupby('Type').size().plot(kind='bar')
```

```python
# BDI histogram for each type
df_resnet.groupby('Type')['BDI'].plot(kind='hist', alpha=0.5, legend=True, bins=30)
```

```python
# Age histogram for each type
df_resnet.groupby('Type')['Age'].plot(kind='hist', alpha=0.5, legend=True, bins=30)
```

```python
# chack the first 25 features histogram for each type on a 5x5 grid
fig, axs = plt.subplots(5, 5, figsize=(20, 20))
for i, ax in enumerate(axs.flat):
    df_resnet.groupby('Type')[feature_cols[i]].plot(kind='hist', alpha=0.5, legend=True, bins=30, ax=ax)
    ax.set_title(feature_cols[i])

plt.tight_layout()
plt.show()
```

```python
# show the first 25 features for each sex on a 5x5 grid
fig, axs = plt.subplots(5, 5, figsize=(20, 20))
for i, ax in enumerate(axs.flat):
    df_resnet.groupby('Sex')[feature_cols[i]].plot(kind='hist', alpha=0.5, legend=True, bins=30, ax=ax)
    ax.set_title(feature_cols[i])

plt.tight_layout()
plt.show()
```
