```python
import pandas as pd
```

```python
# Put the 4 txts and the raw data in the same folder
data_folder = "data"
excel_path = "data/de_hc_adatb√°zis.xlsx"
```

```python
dfs = pd.read_excel(excel_path, sheet_name='Munka1')
metadata_column_names = ["File name", "Type", "Sex", "Age", "BDI"]
excel_filtered = dfs.loc[:,metadata_column_names]
```

```python
ecapa1 = pd.read_csv(f"{data_folder}/dep_merged_output_ecapa.txt", sep="\t", header=None, decimal=',')
ecapa2 = pd.read_csv(f"{data_folder}/hc_merged_output_ecapa.txt",sep="\t", header=None, decimal=',')
ecapa = pd.concat([ecapa1,ecapa2])
ecapa
ecapa['File name'] = ecapa[0].str.split('.', expand=True)[0]
ecapa.drop(columns=[0], inplace=True)
df_ecapa = pd.merge(excel_filtered, ecapa, on='File name', how='inner')
df_ecapa
```

```python
resnet1 = pd.read_csv(f"{data_folder}/dep_merged_output_resnet.txt", sep="\t", header=None, decimal=',')
resnet2 = pd.read_csv(f"{data_folder}/hc_merged_output_resnet.txt",sep="\t", header=None, decimal=',')
resnet = pd.concat([resnet1,resnet2])
resnet['File name'] = resnet[0].str.split('.', expand=True)[0]
resnet.drop(columns=[0], inplace=True)
df_resnet = pd.merge(excel_filtered, resnet, on='File name', how='inner')
df_resnet
```

```python
xvector = pd.DataFrame()
xvectorlist = []
for name in excel_filtered["File name"]:
    a = pd.DataFrame()
    if "DE" in name:
        a = pd.read_csv(f"{data_folder}/raw_dep/{name}.wav_embeddings.txt",sep="\t", header=None, decimal=',')
    else:
        a = pd.read_csv(f"{data_folder}/raw_hc/{name}.wav_embeddings.txt",sep="\t", header=None, decimal=',')
    a['File name'] = name
    xvectorlist.append(a)
xvector = pd.concat(xvectorlist)
xvector.reset_index(drop=True)
df_xvector = pd.merge(excel_filtered, xvector, on='File name', how='inner')
df_xvector
```

```python
print(f"Resnet:{df_resnet.shape}")
print(f"Ecapa: {df_ecapa.shape}")
print(f"XVector: {df_xvector.shape}")
df_xvector.columns
```

```python
import matplotlib.pyplot as plt

def femaleOrMale(text, df, plot):
    females = df[df['Sex']=='F']
    males = df[df['Sex']=='M']
    columns = [item for item in df.columns if item not in metadata_column_names]
    f_mean = females[columns].mean()
    m_mean = males[columns].mean()
    diff = f_mean - m_mean
    max_col = diff.idxmax()
    plot.hist(males[max_col], 20, label="Male",alpha=0.8)
    plot.hist(females[max_col], 20, label="Female",alpha=0.8)
    plot.legend()
    plot.set_title(f"Male or Female: {text} - {max_col}")
    
```

```python
import matplotlib.pyplot as plt

def femaleOrMaleMedian(text, df, plot):
    females = df[df['Sex']=='F']
    males = df[df['Sex']=='M']
    columns = [item for item in df.columns if item not in metadata_column_names]
    f_mean = females[columns].median()
    m_mean = males[columns].median()
    diff = f_mean - m_mean
    max_col = diff.idxmax()
    plot.hist(males[max_col], 20, label="Male",alpha=0.8)
    plot.hist(females[max_col], 20, label="Female",alpha=0.8)
    plot.legend()
    plot.set_title(f"Male or Female: {text} - {max_col}")
```

```python
def depressedOrHealthy(text, df, plot):
    depressed = df[df['Type']=='DE']
    healthy = df[df['Type']=='HC']
    columns = [item for item in df.columns if item not in metadata_column_names]
    d_mean = depressed[columns].mean()
    h_mean = healthy[columns].mean()
    diff = d_mean - h_mean
    max_col = diff.idxmax()
    plot.hist(healthy[max_col], 20, label="Healthy",alpha=0.8)
    plot.hist(depressed[max_col], 20, label="Depressed",alpha=0.8)
    plot.legend()
    plot.set_title(f"Depressed or Healthy: {text} - {max_col}")
```

```python
def ageToHealthy(text, df, plot):
    depressed = df[df['Type']=='DE']
    healthy = df[df['Type']=='HC']
    plot.hist(healthy["Age"], 20, label="Healthy",alpha=0.8)
    plot.hist(depressed["Age"], 20, label="Depressed",alpha=0.8)
    plot.legend()
    plot.set_title(f"Depressed or Healthy: {text} - Age")
```

```python
def completeAnalisis(text, df):
    fig, axes = plt.subplots(2, 2)
    plt.tight_layout()
    femaleOrMale(text, df, axes[0][0])
    depressedOrHealthy(text, df, axes[0][1])
    femaleOrMaleMedian(text, df, axes[1][0])
    ageToHealthy(text, df, axes[1][1])
```

```python
completeAnalisis("Xvector", df_xvector)
```

```python
completeAnalisis("Ecapa", df_ecapa)
```

```python
completeAnalisis("Resnet", df_resnet)
```

```python
def normalize(df):
    df_normalized = df.copy()
    columns = [item for item in df.columns if item not in metadata_column_names]
    df_normalized[columns] = df_normalized[columns].apply(lambda x: (x - x.min()) / (x.max() - x.min()), axis=1)
    return df_normalized
```

```python
ecapa_normalized = normalize(df_ecapa)
xvector_normalized = normalize(df_xvector)
resnet_normalized = normalize(df_resnet)
```

```python
completeAnalisis("Xvector - n", xvector_normalized)
```

```python
completeAnalisis("Ecapa - n", ecapa_normalized)
```

```python
completeAnalisis("Resnet - n", resnet_normalized)
```
